```python
X=data[['Gender','Age','openness','neuroticism','conscientiousness','agreeableness','extraversion']]
y=data[['Personality (Class label)']]
mul_lr = linear_model.LogisticRegression(multi_class='multinomial', solver='newton-cg',max_iter =1000)
mul_lr.fit(X, y)

testdata =pd.read_csv('test.csv')
for i in range(len(testdata['Gender'])):
    if testdata['Gender'][i]=="Male":
        testdata['Gender'][i]=1
    else:
        testdata['Gender'][i]=0
X_test=testdata [['Gender','Age','openness','neuroticism','conscientiousness','agreeableness','extraversion']]
y_test=testdata [['Personality (class label)']]

y_pred=mul_lr.predict(X_test)


accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.ensemble import AdaBoostClassifier
knn = AdaBoostClassifier()
knn.fit(X, y)

y_pred = knn.predict(X_test)

confusion = metrics.confusion_matrix(y_test, y_pred)
print("\n\n\n")
print(confusion)
print(metrics.classification_report(y_test, y_pred))
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy of model: {}%".format(accuracy*100))
```

```python
from sklearn.ensemble import RandomForestClassifier
knn = RandomForestClassifier()
knn.fit(X, y)

y_pred = knn.predict(X_test)

confusion = metrics.confusion_matrix(y_test, y_pred)
print("\n\n\n")
print(confusion)
print(metrics.classification_report(y_test, y_pred))
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy of model: {}%".format(accuracy*100))
```
